```python
search = data[['identity','qaccver','saccver']]
search.dropna(inplace=True)
search.head()
```

```python
search['qs']=search['qaccver'] + search['saccver']
search['sq']=search['saccver'] + search['qaccver']
search.head()
```

```python
import numpy as np
r7 = r5.groupby(['g1','g2','g2t','pd1','pd2']).agg(cont =('eentity','size'),dic=('eentity',pd.Series.nunique))
r7.reset_index(inplace=True)
r7.head()
```

```python
len(r7)
```

```python
r7['qs']=r7['pd1'] + r7['pd2']
r7.head()
```

```python
r8 = r7[(r7['qs'].isin(search['qs'].values))|(r7['qs'].isin(search['sq'].values))]
len(r8)
```

```python
res1 = pd.merge(r8,search[['qs','identity']],left_on='qs',right_on='qs')
len(res1)
```

```python
res2 = pd.merge(r8,search[['sq','identity']],left_on='qs',right_on='sq')
res2.head()
```

```python
len(res2)
```

```python
res = pd.concat([res1[['g1','g2','g2t','pd1','pd2','cont','dic','identity']],res2[['g1','g2','g2t','pd1','pd2','cont','dic','identity']]])
len(res)
```

```python
res.head()
```

```python
res['score']=res['cont']*res['dic']*res['identity']
res.head()
```

```python
len(res)
```

```python
res3 = res[res['score']>=502.36]
len(res3)
```
